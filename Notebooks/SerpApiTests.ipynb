```python
API_KEY = str(open('../.serpapikey', 'r').readlines()[0]).rstrip()
```

```python
from serpapi import GoogleSearch

params = {
  "api_key": API_KEY,
  "engine": "google",
  "q": "biobiochile",
  "location": "Chile",
  "google_domain": "google.cl",
  "gl": "cl",
  "hl": "es",
  "safe": "active",
  "tbm": "nws",
  "num": "200"
}

search = GoogleSearch(params)
results = search.get_dict()
```

```python
for i in results['news_results']:
    if 'horas' in i['date']:
        print('##',i['title'],'##')
```

```python
for i in results['news_results']:
    if 'horas' in i['date']:
        if '.cl' in i['link'] or '.com' in i['link']:
            print('##',i['title'],'##')
        else:
            print('!!',i['title'],'!!')
```
